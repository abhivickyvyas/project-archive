```python
import random
# random.sample(the_list, 50)
def find_initial_seed_points(all_data_dict,K):
  list_of_keys=list(all_data_dict.keys())
  random_seeds=random.sample(list_of_keys,K)
  random_seeds_vec_dict={}
  for k in random_seeds:
    random_seeds_vec_dict[k]=all_data_dict[k]
  return random_seeds_vec_dict
```

```python
def FindEuclidDist(vec1,vec2):
    sum=0
    # for i in range(len(vec1)):
    #     sum+=(vec1[i]-vec2[i])**2
    sum=np.sum((np.array(vec1)-np.array(vec2))**2,axis=0)
    eucdis=math.sqrt(sum)
    return eucdis
```

```python
def find_clusters_list(seeds,all_data_dict):
    for x,y in all_data_dict.items():
      minis=100
      min_key=""
      # print(x,y,'\n\n====')
      for skey,svec in seeds.items():
          # print(skey,svec['vector'])
          seed_vec_ecu=FindEuclidDist(y,svec['vector'])
          if(seed_vec_ecu<minis):
             minis=seed_vec_ecu
             min_key=skey
      # print(x,min_key,"\n")      
      seeds[min_key]['cluster_list'].append(x)
      # print(x,seeds[min_key]['cluster_list'])  
      # # print(seeds,min_key,maxis)
      # # break
    return seeds

```
